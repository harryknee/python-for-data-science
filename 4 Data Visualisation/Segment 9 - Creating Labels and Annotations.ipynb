## Labelling plot features

### The functional method

```python
x = range(1, 10)
y = [1, 2, 3, 4, .5, 4, 3, 2, 1]
plt.bar(x, y)

plt.xlabel('your x-axis label')
plt.ylabel('your y-axis label')
```

```python
z = [1,2,3,4,.5]
veh_type = ['bicycle', 'motorbike', 'car', 'van', 'stroller']

plt.pie(z, labels = veh_type)
plt.show()
```

### The object-oriented method

```python
address = 'C:/Users/zxgsh/Desktop/DS/mtcars.csv'
cars = pd.read_csv(address)

mpg = cars.mpg

fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

mpg.plot()
plt.grid()

# label
ax.set_xticks(range(32))

ax.set_xticklabels(cars.model, rotation = 60, fontsize = 'medium')
ax.set_title('Miles per gallon of cars in mtcars dataset')

ax.set_xlabel('car model')
ax.set_ylabel('miles/gal')

# annotation
ax.legend(loc = 'best')

ax.set_ylim([0, 45])

ax.annotate('Toyota Corolla', xy = (19, 33.9), xytext = (21, 35), arrowprops = dict(facecolor = 'black', shrink = 0.05))
```
