```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
# print matplotlib plot inline in jupyter
%matplotlib inline
rcParams['figure.figsize'] = 5, 4 # 5-inch wide, 4-inch tall
```

## Defining axes, ticks and grids

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1]) # axes argument: rect = [left, bottom, width, height]

ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

# setting limits
ax.set_xlim([1, 9])
ax.set_ylim([0, 5])

# setting ticks
ax.set_xticks(range(1, 11))
ax.set_yticks(range(0, 6))

ax.grid()
ax.plot(x, y)
```

## Subplot

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(x)
ax2.plot(x, y)
#two plots are presented inline
```
