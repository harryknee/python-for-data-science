```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
```

```python
import seaborn as sb
sb.set_style('whitegrid')
```

## Explore dataset distributions with histograms

```python
address = 'C:/Users/zxgsh/Desktop/DS/mtcars.csv'
cars = pd.read_csv(address)

cars.index = cars.model

mpg = cars['mpg']

mpg.plot(kind = 'hist')
```

```python
plt.hist(mpg)
plt.plot()
```

```python
sb.histplot(mpg)
```

## Seeing scatterplots in action

```python
cars.plot(kind = 'scatter', x = 'hp', y = 'mpg', c = ['darkgrey'], s = 150)
```

```python
sb.regplot(x = 'hp', y = 'mpg', data = cars, scatter = True) 
# scatter parameter shows the datapoints, default = True
```

## Generating a scatter plot matrix

```python
sb.pairplot(cars)
```

```python
cars_subset = cars[['mpg', 'disp', 'hp', 'wt']]
sb.pairplot(cars_subset)
plt.show()
```

## Building boxplots

```python
cars.boxplot(column = 'mpg', by = 'am')
cars.boxplot(column = 'wt', by = 'am')
```

```python
sb.boxplot(x = 'am', y = 'mpg', data = cars, palette = 'hls')
```
